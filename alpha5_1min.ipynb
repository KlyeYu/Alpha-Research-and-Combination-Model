```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import Tools
```

```python
path = r'PTA_data_day'
dataList = []
IC = []
for filename in os.listdir (path):
    if filename == "ta1_day_20220902.csv":
        break
    head = r"PTA_data_day/"
    data = pd.read_csv(head + filename)
    groupby = Tools.changeFreq(data, "1min")
    returnData = Tools.extractReturn(groupby.apply(Tools.lastPrice))
    mbarate = groupby.apply(Tools.maxBidAskRate)
    frame = pd.concat([returnData, mbarate], axis = 1)
    dataList.append(frame)
    IC.append(frame.corr().iloc[1, 0])
alpha5Return = pd.concat(dataList, axis = 0)
```

```python
plt.plot(range(len(IC)), IC)
plt.xlabel("Time")
plt.ylabel("IC")
```

```python
alpha5Return.dropna(inplace = True)
```
